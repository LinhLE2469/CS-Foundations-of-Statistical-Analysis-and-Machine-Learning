```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn 
import seaborn as sns
```

```python
dataset = pd.read_csv('Titanic.csv')
```

```python
dataset.head()
```

```python
# let's see number of row and column
dataset.shape
```

```python
# Let's see number of person in Titanic
dataset['PassengerId'].count()
```

```python
# Let's see distribution of age in this dataset
dataset['Age'].describe()
# dataset.Age.describe()--> difference way to show 
```

```python
# plt.hist(dataset['Age'],bins=40,alpha=1, rwidth=0.9)
plt.figure(figsize=[8,5])
dataset.Age.hist(bins=40,alpha=1, rwidth=0.9)
plt.title('Age distribution', fontsize = 15)
plt.xlabel('Age', fontsize = 15)
plt.ylabel('Number of person', fontsize = 15)
plt.show()
```

From this histogram. We can see that many children is below 5 year olds. Some of them are babies which we see that there is a peak around 1 year. The passenger's age is distributed around from 20 to 30 year olds. 

```python
# Let's see Gender in Titanic
# dataset['Sex'].value_counts()

plt.figure(figsize = [15, 5])
plt.subplot(1, 2, 1)
Sex=dataset.groupby(['Sex']).size()
Sex.plot(kind='bar',color=['Pink', 'blue'])
plt.ylabel('Count')
plt.title('Gender Split')
plt.grid(axis='y', alpha=0.55)
plt.subplot(1, 2, 2)
x=dataset.groupby(['Sex']).size()

plt.pie(x, labels=['Fmale','Male'],autopct='%1.1f%%',textprops=dict(color="white",weight='bold'));
plt.legend(loc='center right', bbox_to_anchor=(1,0.3,0.4,1))
plt.title('Gender Split')
```

```python
# Let's see Average fare for different SES
plt.figure(figsize = [15, 5])
plt.subplot(1, 2, 1)
Pclass=dataset.groupby(['Pclass']).size()
Pclass.plot(kind='bar',color=['Blue', 'orange','green'])
plt.ylabel('Count')
plt.title('Pclass distribution')
plt.grid(axis='y', alpha=0.55)
plt.subplot(1, 2, 2)
# Average Fare based on Pclass
# dataset.groupby('Pclass')['Fare'].mean()
sns.barplot(x="Pclass",y="Fare",data=dataset,estimator=np.mean)
plt.ylabel("Average Fare")
plt.xlabel("Socio-Econmic Status")
plt.title("Average fare for different SES")
plt.grid(axis='y', alpha=0.55)
```

```python
# Age distribution by embarkation
# dataset.groupby('Embarked').Age.describe()
plt.figure(figsize = [15, 5])
plt.subplot(1, 2, 1)
Embarked=dataset.groupby(['Embarked']).size()
Embarked.plot(kind='bar',color=['Blue', 'orange','green'])
plt.ylabel('Count')
plt.title('Embarked Atribution')
plt.grid(axis='y', alpha=0.55)
plt.subplot(1, 2, 2)
p = sns.violinplot(data = dataset, x = 'Embarked', y = 'Age',order=["S","C","Q"])
p.set(title = 'Age Distribution by Embarkation Point', xlabel = 'Location', ylabel = 'Age Distribution', xticklabels = ['Southampton','Cherbourg', 'Queenstown']);
plt.grid(axis='y', alpha=0.55)
```

```python
# Check free ticket with sort method
# (dataset['Fare']==0).sum()
dataset[dataset['Fare']==0].sort_values(by=['Ticket']) 
```

# Survived Information

```python
# Lets see the age distribution by survival
plt.figure(figsize=[8,5])
Age_die=dataset['Age'][dataset['Survived']==0]
Age_survived=dataset['Age'][dataset['Survived']==1]
plt.hist([Age_die,Age_survived],stacked= False , color=['black', 'Green'],alpha=0.5, rwidth=0.9)
plt.title('Age distribution by survival', fontsize = 15)
plt.xlabel('Age', fontsize = 15)
plt.ylabel('Number of person', fontsize = 15)
plt.legend(['Died','Survived'], fontsize = 15)
plt.grid(axis='x', alpha=0.55)
plt.show()
```
